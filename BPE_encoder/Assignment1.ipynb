## Task 2. Implement [Byte-Pair Encoding(BPE)](https://arxiv.org/pdf/1508.07909.pdf) Tokenizer (80 points)

### Task 2.1. Implementation (60 points)

Implement the tokenizer as the BPETokenizer class:
* Implement `train` method that learns merges and builds the vocabulary of the specified `vocab_size` (25 points).
* Implement `tokenize` method that should tokenize the text according to the learnt merges (25 points).

Your code should have docstrings and comments (10 points).

```python
from typing import List
from collections import Counter, defaultdict
import re
from tqdm import tqdm
import string
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()

class BPETokenizer:

    def __init__(self, vocab_size: int) -> None:
        '''
        Parameters:
        vocab_size (int) - the size of vocabulary we want to achieve. Algorithm stops training when reaches it.
            Set it 'no' if no extra merges needed (so the vocabuary will only contain every single character from corpus)
            
        '''
        self.vocab_size = vocab_size

    def get_freqs_of_tokens_pairs(self, word_freq : Counter) -> dict:
        '''
        This function calculates the frequency of pairs of tokens that we currently have determined

        Parameters:
        c (Counter) - the counter of words in the corpus. Every word in keys should have the form of tokens separated by spaces.
            For example, if we have dictionary ['a', 'b', 'cd'], then the word 'acdb' should be represented as 'a cd b'

        Returns:
        The dictionary of occurences of pairs, where key is the tuple of two characters, the value is the frequency.
            For example, this dictionary can look like {('ab', 'NO') : 5, ('a','<\w>') : 2}
        '''
        
        pairs = defaultdict(int) # Here for every new pair of tokens we have default value of 0 (zero occurences)
        
        for word, freq in word_freq.items(): #iterate over words and their frequencies
            tokens = word.split() # split over tokens (they are separated by space)
            char_zip = list(zip(tokens, tokens[1:])) # Here we create list of pairs of tokens
            for char in char_zip: # Iterating over pairs
                pairs[char] += freq # update frequency of every pair of tokens.
                
        return pairs 

    def merge_byte_pairs(self, best_pair : tuple, word_freq) -> dict:
        '''
        This function merges two tokens together

        Parameters:
        best_pair (tuple) - the tuple with two string tokens to be merged. For example: ('a', 'b')
        word_freq (Counter or dict) - the Counter or dict of words in the corpus. Reminder, every word is the tokens separated by spaces

        Returns:
        The dictionary with words and their frequences. Updated version from parameters with words that account for new tokens.
        '''
        best_pair_space = " ".join(best_pair)
        best_pair_merged = "".join(best_pair)
        merged_dict = {} # new dictionary that stores frequency of tokenized words
        for word in word_freq: # Iterating over all keys in the dictionary of frequencies of words in the corpus
            word_merged = word.replace(best_pair_space, best_pair_merged) # replace two tokens by one
            merged_dict[word_merged] = word_freq[word] # save new word with same frequency as it had
        return merged_dict

    def calculate_tokens_frequencies(self, word_freq) -> dict:
        '''
        This function calculates the frequency of every individual token

        Parameters:
        word_freq (Counter or dict) - the frequencies of words in the corpus. Reminder, every word is the tokens separated by spaces

        Returns: 
        The dictionary with tokens and their frequencies.
        '''
        token_freq = defaultdict(int)
        for word, freq in word_freq.items(): 
            chars = word.split() # split the text over tokens by spaces
            for char in chars: # update frequency of the token by frequency of the word itself
                token_freq[char] += freq 
                
        return token_freq
        
    def train(self, corpus: List[str]) -> None:
        '''
        This is the function that learns the vocabulary of the tokenizer by merges of most frequent combination of characters into tokens

        Parameters:
        corpus (List[str]) - the list with the texts.
        '''
        word_freq = Counter([])

        # preprocessing. 
        for text in corpus:
            text = text.lower().replace("\n", " ")
            tokens = text.split() # splitting the text over words
            tokens = [" ".join(tok) + ' </w>' for tok in tokens] # splitting words over characters. also, adding special symbol of word-end
            word_freq.update(tokens)

        # If we do not need to do any merges => use only standart vocabulary of characters
        if self.vocab_size == 'no':
            subword_tokens = self.calculate_tokens_frequencies(word_freq)
            self.vocab = list(subword_tokens.keys())
            return
            
        for _ in range(self.vocab_size): 
            # Here we searching for the most frequent pair of tokens and merge them
            pair_freqs = self.get_freqs_of_tokens_pairs(word_freq) # calculating frequencies of tokens
            best_pair = max(pair_freqs, key = pair_freqs.get) # searching for the most frequent pair
            word_freq = self.merge_byte_pairs(best_pair, word_freq) # merging
            subword_tokens = self.calculate_tokens_frequencies(word_freq) # getting the tokens list + their frequencies
            
            if len(subword_tokens) >= self.vocab_size: #stop criteria reached
                break
        self.vocab = list(subword_tokens.keys()) # saving the tokens as vocabulary
        

    def tokenize(self, text: str) -> List[str]:
        '''
        This function tokenizes the input text into tokens learnt from training.
        Be careful with unseen data. Especially, when switching the language.

        Parameters:
        text (str) - the text to tokenize.

        Returns:
        The list of tokens for the given text
        '''
        words = text.lower().replace("\n", " ").split() # preprocessing
        subwords_list = []
    
        for word in words:
            start_index = 0
        
            # continue until the start index is beyond the length of the word
            while start_index < len(word):
                longest_match = ''
            
                # iterate over the vocabulary to find the longest matching subword
                for subword in self.vocab:
                    if word.startswith(subword, start_index):
                        if len(subword) > len(longest_match): # we are interested in longest subwords
                            longest_match = subword
            
                if longest_match: # i a match was found, add it to the list and move to the next part of the word
                    subwords_list.append(longest_match)
                    start_index += len(longest_match)
                else:  # if no match was found, add the current character as a subword. 
                    # this part triggers only on unseen data when algorithm sees characters it has not seen furing training
                    
                    subwords_list.append(word[start_index])
                    start_index += 1
        return subwords_list
```

### Task 2.2. Analysis on Tweets Dataset (10 points)

Train the BPE tokenizer on the tweets dataset. Try to tokenize the tweets with the tokenizer of different `vocab_size`. For example, train the BPE tokenizer with `vocab_size` of [base_vocab_size, 250, 500, 750, 1000]. Plot the dependency of the average length of the tokenized tweet by `vocab_size` to analyze how `vocab_size` affects the length of the tokenized tweet on average. Tell what `vocab_size` is preferrable and why.

```python
with open("./tweets/file1") as f1, open("./tweets/file2") as f2, open("./tweets/file3") as f3, open("./tweets/file4") as f4, open("./tweets/file5") as f5:
    all_tweets = f1.readlines() + f2.readlines()+ f3.readlines()+ f4.readlines()+ f5.readlines()
```

```python
vocab_sizes = ['no', 250, 500, 750, 1000, 2000]
avgs = []
for s in vocab_sizes:
    tokenizer = BPETokenizer(s)
    tokenizer.train(all_tweets)
    
    sum = 0
    for tweet in all_tweets:
        sum += len(tokenizer.tokenize(tweet))
    sum /= len(all_tweets)
    avgs.append(sum)
```

```python
plt.plot([141, 250, 500, 750, 1000, 2000], avgs)
plt.title("Average tokens for tweet")
plt.xlabel("Vocabulary size")
plt.ylabel("Average number of tokens for tweet")
```

### Conclusion
We see vocabulary size decreases almost exponentially the average number of tokens. The average number of tokens decreases because when algorithm uses higher vocabulary size it remembers more words entirely and more different subwords together. Therefore, in average we need less tokens for word. We can choose the vocabulary size that ends the rapid decreasing or that size that gives the least average length before plato.\
In my opinion, it is reasonable here to choose vocabulary size of 2000, because decreasing is still fast here, and less average tokens per words is better.

### Dataset

```python
import pandas as pd
rus_poems = pd.read_csv("./tweets/poems.csv", encoding="utf-8")
rus_poems = rus_poems.dropna().text
```

```python
vocab_sizes = ['no', 250, 500, 750, 1000, 2000]
avgs = []
for s in tqdm(vocab_sizes):
    tokenizer = BPETokenizer(s)
    tokenizer.train(rus_poems[:400])
    
    sum = 0
    div = 0
    for poem in rus_poems[:400]:
        for sent in poem.split("."):
            sum += len(tokenizer.tokenize(sent))
            div+=1
    sum /= div
    avgs.append(sum)
    
plt.plot([110, 250, 500, 750, 1000, 2000], avgs)
plt.title("Average tokens for sentence in russian poem ")
plt.xlabel("Vocabulary size")
plt.ylabel("Average number of tokens per sentence")
```

### Conclusion
Here we also see something exponential\
We see that number of tokens for different vocabulary size is different for english tweets and sentences in russian poems. In case of russian poems, we have more tokens. The reasons are the following:
1. Russian language has more different forms of the same words, so when in english we can memorize the entire word in two tenses, in russian it takes a lot of tokens to memorize words in different tenses, sexes and cases. So, we memorize a lot of small pieces like suffixes, prefixes, endings. As a result, every word takes a lot of tokens for different parts of words
2. Tweets are much shorter than sentences in poems. In poem, writer tries to express his feelings and thoughts in detail, whereas writers of tweets usually do not spend much time on writing, and do not use a lot of words to say what they want.
